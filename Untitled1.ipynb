Quel Maire a planté le plus d’arbres à Grenoble ?

```python
mayors = [['Paul Mistral',1919],['Leon Martin',1932],['Paul Cocat',1935],['Frederic Lafleur',1944],['Leon Martin',1945],['Marius Bally',1947],['Raymond Perinetti',1948],['Leon Martin',1949],['Albert Michallon',1959],['Hubert Dubedout',1965],['Alain Carignon',1983],['Michel Destot',1995],['Eric Piolle',2014]] 
```

Avec cette liste de maires avec leur année d'élection on va construire un datatype contenant les années et le maire en poste.

```python
def mayor_of_year(year):
    for mandat in mayors:
        if year >= mandat[1]:
            mayor = mandat[0]
    return(mayor)

data_mayor = []
for year in range(1924, 2020):
    data_mayor.append((year, mayor_of_year(year)))
df2 = pd.DataFrame(data_mayor ,columns=['ANNEEDEPLANTATION','mayor'])
df2
```

On fait une jointure des deux data frames, avec comme clé l'année de plantation :

```python
left = df
right = df2

result = pd.merge(left, right, on='ANNEEDEPLANTATION').head(50)
result[['GENRE_BOTA', 'mayor']]
```
