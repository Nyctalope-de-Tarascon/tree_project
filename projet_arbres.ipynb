Mise en forme des donn√©es pour obtenir une liste de dicos

```python
import csv
import matplotlib.pyplot as plt
import timeit

mayors = [['Paul Mistral',1919],['Leon Martin',1932],['Paul Cocat',1935],['Frederic Lafleur',1944],['Leon Martin',1945],['Marius Bally',1947],['Raymond Perinetti',1948],['Leon Martin',1949],['Albert Michallon',1959],['Hubert Dubedout',1965],['Alain Carignon',1983],['Michel Destot',1995],['Eric Piolle',2014]] 
arbres = []
headers = []
with open("Data/ESP_PUBLIC.IDENTITE_ARBRE.csv", newline='') as csvfile:
    lines = csv.reader(csvfile)
    i = 0
    for row in lines :
        if i == 0:
            for header in row:
                headers.append(header)
        else:
            arbres.append(row)
        i += 1
big_data = []
j = 0
for arbre in arbres:
    big_data.append({})
    k = 0
    for element in headers:
        big_data[j][element] = arbre[k]
        k += 1
    j += 1
  
```

Ecriture d'un fichier

```python
with open("my_super_new_file.csv", mode='w') as new_csv_file:
    my_csv_writer = csv.writer(new_csv_file, delimiter='|')
    
    my_csv_writer.writerow(headers)
    for arbre in arbres:
        my_csv_writer.writerow(arbre)
```

Fonction qui renvoie le nombre d'arbres plant√©s pendant une ann√©e donn√©e en argument

```python
def tree_of_year(year):
    tree_of_year = 0
    for arbre in big_data:
        if arbre["ANNEEDEPLANTATION"] != "":
            if int(arbre["ANNEEDEPLANTATION"]) == year:
                tree_of_year += 1
    return tree_of_year
```

Fonction qui renvoie le nom du maire pour une ann√©e donn√©e en argument

```python
def mayor_of_year(year):
    for mandat in mayors:
        if year >= mandat[1]:
            mayor = mandat[0]
    return(mayor)
```

Fonction qui renvoie une liste de tuples contenant le genre botanique et l'ann√©e de plantation de chaque arbre

```python
def year_vs_botanic_type():
    year_and_type = []
    for arbre in big_data:
        year_and_type.append((arbre["GENRE_BOTA"], arbre["ANNEEDEPLANTATION"]))
    return(year_and_type)
```

Fonction qui renvoie le nombre d'arbres qui n'ont pas de genre botanique

```python
def missing_genre_bota():
    count = 0
    for genre in year_vs_botanic_type():
        if genre[0] == "":
            count += 1
    return count
```

calcule le temps d'ex√©cution du recensement de tous les genres botaniques pr√©sents dans les donn√©es

```python
def get_all_genres():
    genre_list = []
    for arbre in big_data:
        if arbre["GENRE_BOTA"] not in genre_list and arbre["GENRE_BOTA"] != "":
            genre_list.append(arbre["GENRE_BOTA"])
    return (len(genre_list))

timeit.timeit(get_all_genres, number=1)
```

Identification de toutes les esp√®ces :

```python
def get_all_species_2():
    species_list = set((arbre["GENRE_BOTA"],arbre["ESPECE"]) for arbre in big_data if arbre["ESPECE"] != "")
    return species_list

get_all_species_2()
```

On it√®re pour trouver le nombre d'arbres de chaque genre botanique

```python
def number_of_genre_botanic():
    number_of_gb = {}
    for genre_botanic in get_all_genres_2():
        number_of_gb[genre_botanic] = 0
    for arbre in big_data:
        if arbre["GENRE_BOTA"] in number_of_gb:
            number_of_gb[arbre["GENRE_BOTA"]] += 1
    return number_of_gb
print(number_of_genre_botanic())
```

```python
On cherche les arbres uniques repr√©sentants de leur genre :
```

```python
def orphant_trees():
    orphants = []
    for key, val in number_of_genre_botanic().items():
        if val == 1:
            orphants.append(key)
    return orphants
print(orphant_trees())
```

On va commencer par arranger le set de liste d'esp√®ces en dict de dict (cl√©s : GENRE puis ESPECE) avec un nombre de repr√©sentants initialis√© √† 0

```python
def all_species_in_genre():
    empty_specie_count = {}
    for genre_specie in get_all_species_2():
        if genre_specie[0] in empty_specie_count:
            empty_specie_count[genre_specie[0]][genre_specie[1]] = 0
        else:
            empty_specie_count[genre_specie[0]] = {genre_specie[1]: 0}
    return empty_specie_count
species_table = all_species_in_genre()
```

Il suffit d'it√©rer sur les arbres pour incr√©menter les totaux

```python
def number_of_genre_species():
    for arbre in big_data:
        if arbre["GENRE_BOTA"] in species_table and arbre["ESPECE"] in species_table[arbre["GENRE_BOTA"]]:
            species_table[arbre["GENRE_BOTA"]][arbre["ESPECE"]] += 1
    return species_table
number_of_genre_by_species = number_of_genre_species()
```
